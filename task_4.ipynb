```python
# Simple List
l1_alpha = ['a', 'b', 'c']

# Range List (1 to 9)
l1_range = list(range(1, 10))

# Range with Step (1 to 50, jumping by 5)
l1_step = list(range(1, 50, 5))

# Nested List (Matrix style)
l1_nested = [['a', 'b'], [1, 2]]

# Deeply Nested List
l1_deep = [list(range(1, 5)), [[1, 2], ['5']]]

print(f"Step List: {l1_step}")
print(f"Deeply Nested: {l1_deep}")
```

##### APPEND

```python
l1 = ['a', 'b', 'c']
l1.append('d')
print(l1)
```

##### EXTEND

```python
l1 = [1, 2, 3]
l2 = [4, 5, 6]
l1.extend(l2)
print(l1)
```

##### INSERT

```python
l1 = ['a', 'c']
l1.insert(1, 'b') # Inserts 'b' at index 1
print(l1)
```

##### REMOVE

```python
l1 = [10, 20, 30, 20]
l1.remove(20)
print(l1) #(Only the first 20 is gone)
```

##### POP

```python
l1 = ['apple', 'banana', 'cherry']
last_item = l1.pop()
print(last_item, l1) 
# last_item is 'cherry', l1 is now ['apple', 'banana']
```

##### CLEAR

```python
l1 = [1, 2, 3, 4, 5]
l1.clear()
print(l1)
```

##### INDEX

```python
l1 = ['a', 'b', 'c', 'd']
position = l1.index('c')
print(position)
```

##### COUNT

```python
l1 = [1, 2, 2, 3, 2, 4]
how_many = l1.count(2)
print(how_many)
```

##### SORT

```python
l1 = [40, 10, 30, 20]
l1.sort()
print(l1)
```

##### REVERSE

```python
l1 = [1, 2, 3]
l1.reverse()
print(l1)
```
